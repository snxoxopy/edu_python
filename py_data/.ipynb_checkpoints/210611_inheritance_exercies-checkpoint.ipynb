```python
import random
import turtle
from abc import *



class Duck(metaclass=ABCMeta): 

    DUCK_SIZE = 30 # class
    
   
    def  __init__(self, turtle):
        self._x = random.randint(-300,300) #instance
        self._y = random.randint(-300,300)
        self._turtle = turtle
        
    @abstractmethod #abstract method           
    def display(self):
        pass

        
    def screen_reset(self):
        self._turtle.reset()


    def quack(self):
        #self._turtle.color('black')
        self._turtle.penup()
        self._turtle.setpos(self._x + 40, self._y+40)
       
        self._turtle.write("quack")  #print text
        self._turtle.setpos(self._x, self._y)
        self._turtle.pendown()
        
    
    def move(self):
        #self._turtle.color('black')
        print(self._x, ',', self._y, ':', Duck.DUCK_SIZE)  
        self._turtle.penup()
        self._turtle.setpos(self._x - 50, self._y+40)
        
        self._turtle.write("move")  #print text
        self._turtle.setpos(self._x, self._y)
        self._turtle.pendown()

        
class MallardDuck(Duck): #ingeritance
    
    def  __init__(self, turtle):
        super().__init__(turtle)
        self.screen_reset()
        self._turtle.color('blue')
        
        
    # Method's Override
    def display(self):
        self._turtle.begin_fill()
        self._turtle.circle(Duck.DUCK_SIZE)
        self._turtle.end_fill()
        self._turtle.penup()
        self._turtle.goto(self._x, self._y)
        self._turtle.pendown()
        print(self._x, ',', self._y, ':', Duck.DUCK_SIZE)
    
    

class RedDuck(Duck): #ingeritance

    def __init__(self, turtle):
        super().__init__(turtle)
        self._turtle.color('red')
        
    # Method's Override
    def display(self):
        self._turtle.begin_fill()
        self._turtle.circle(Duck.DUCK_SIZE)
        self._turtle.end_fill()
        self._turtle.penup()
        self._turtle.goto(self._x, self._y)
        self._turtle.pendown()
        print(self._x, ',', self._y, ':', Duck.DUCK_SIZE)
   

 
class DuckManager:
    
    __ist_duck = []
    __instancce = None
       
    def __init__(self):
        if not DuckManager.__instance:
            pass
        else:
            self.getInstance()    
    
    @classmethod        
    def getInstance(cls):
        if not cls.__instance:
            cls.__instance = DuckManager()
        return cls.__instance
       
    def addDuck(self, duck):
        DuckManager.__ist_duck.append(duck)
            
    def displayAllDucks(self):
            for duck in DuckManager.__ist_duck:
                if duck != None:
                    duck.display()
                    duck.move()
                    duck.sound()
            turtle.mainloop()



dm = DuckManager.getInstance()


#for i in range(0, 5):
#    dm.addDuck(MallardDuck(turtle))
#    dm.addDuck(RedDuck(turtle))
#
#dm.displayAllDucks()
#    
```
