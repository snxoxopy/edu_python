```python
import random
from abc import abstractmethod
from abc import ABC
from abc import ABCMeta
import turtle


class Duck(metaclass=ABCMeta):
    #cnt = 0
    
    __DUCK_SIZE = 30
    __DUCK_CNT = 0
    
    def __init__(self, turtle):
        #protected
        self._x = random.randint(-300,300)
        self._y = random.randint(-300,300)
        self._turtle = turtle
        
    def move(self):
        self._turtle.penup()
        self._turtle.setpos(self._x - 50, self._y+40)
        
        self._turtle.write("move")  #print text
        self._turtle.setpos(self._x, self._y)
        self._turtle.pendown()

        print("move")
        
    def quack(self):
        self._turtle.penup()
        self._turtle.setpos(self._x + 40, self._y+40)
       
        self._turtle.write("quack")  #print text
        self._turtle.setpos(self._x, self._y)
        self._turtle.pendown()
        print("quack")
        
    
    @abstractmethod    
    def display(self):
        Duck.__DUCK_CNT += 1
        pass
   
   
    
class MallardDuck(Duck):
    
    def __init__(self, turtle):
        super(MallardDuck, self).__init__(turtle)
        self.__myShape = "MallardDuck"
        
    def display(self):
        #Duck.__DUCK_CNT += 1
        self._turtle.color('blue')
        self._turtle.begin_fill()
        self._turtle.circle(Duck.DUCK_SIZE)
        self._turtle.end_fill()
        self._turtle.penup()
        self._turtle.goto(self._x, self._y)
        self._turtle.pendown()
        self._turtle.penup()
        self._turtle.goto(self._x+39, self._y)
        self._turtle.write(Duck.cnt) 
        self._turtle.pendown()
        print(self._x, ",", self._y, ":", self.__myShape)
        
        
class RedDuck(Duck):
    
    def __init__(self, turtle):
        super(RedDuck, self).__init__(turtle)
        self.__myShape = "RedDuck"
        
    def display(self):
        self._turtle.color('red')
        self._turtle.begin_fill()
        self._turtle.circle(Duck.__DUCK_SIZE)
        self._turtle.end_fill()
        self._turtle.penup()
        self._turtle.goto(self._x, self._y)
        self._turtle.pendown()
        print(self._x, ',', self._y, ':', Duck.__DUCK_SIZE)
        #Duck.__DUCK_CNT += 1
        self._turtle.penup()
        self._turtle.goto(self._x+30, self._y)
        self._turtle.write(Duck.cnt) 
        self._turtle.pendown()
        print(self._x, ",", self._y, ":", self.__myShape)
        
# --- main Test Code ---
#md = MallardDuck()
#md.display()
#
#rd = RedDuck()
#rd.display()
#
```
