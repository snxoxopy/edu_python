```python
import random
from abc import abstractmethod
from abc import ABC
from abc import ABCMeta
import turtle


class Duck(metaclass=ABCMeta):
    #cnt = 0
    
    DUCK_SIZE = 30
    _turtle = None
    #__DUCK_CNT = 0
    
    def __init__(self):
        #protected
        self._x = random.randint(-300,300)
        self._y = random.randint(-300,300)
        #self._turtle = turtle
        #self.screen_reset()
        
    def swim(self):
        self._turtle = turtle
        self._turtle.color("black")
        self._turtle.penup()
        self._turtle.setpos(self._x - 50, self._y + 40)  
        self._turtle.write("수영")  #print text
        self._turtle.setpos(self._x, self._y)
        self._turtle.pendown()
        print("수영")
            
    
    @abstractmethod    
    def display(self):
        #Duck.__DUCK_CNT += 1
        pass
    
   # def screen_reset(self):
    #    self._turtle.reset()
   
        
```

```python
class Quack(metaclass=ABCMeta):
    
    def __init__(self):
        _turtle = None
    

    @abstractmethod   
    def quack(self):
        pass
    
    
class Fly:
    
    def __init__(self):
        _turtle = None
    
    def fly(self):
        self._turtle.color("black")
        self._turtle.penup()
        self._turtle.setpos(self._x - 50, self._y-20)
        self._turtle.write("날다")  #print text
        self._turtle.setpos(self._x, self._y)
        self._turtle.pendown()
        print("날다")
    

    @abstractmethod   
    def fly(self):
        pass
```

```python
class DuckManager:

    __duck_list = []
    __instance = None
    
    def __init__(self):
        #self.makeDucks()
        if not DuckManager.__instance:
            print('creation')
        else:
            print('already', self.getInstance())
            
    @classmethod
    def getInstance(cls):
        if not cls.__instance:
            cls.__instance = DuckManager()
        return cls.__instance

    def makeDucks(self):
        for i in range(0,10):
            type = random.randint(0,3)
            if type == 0:
                DuckManager.__duck_list.append(MallardDuck())
            elif type == 1:
                DuckManager.__duck_list.append(RedDuck())
            elif type == 2:
                DuckManager.__duck_list.append(RubberDuck())
            else:
                DuckManager.__duck_list.append(DecoyDuck())
        
    def displayAllDucks(self):
        for duck in DuckManager.__duck_list:
            if duck != None:
                duck.display()
            
    def swimAllDucks(self):
        for duck in DuckManager.__duck_list:
            duck.swim()
        
    def flyDucks(self):
        for duck in DuckManager.__duck_list:
            if isinstance(duck, Fly):
                duck.fly()
    
    def quackDucks(self):
        for duck in DuckManager.__duck_list:
            if isinstance(duck, Quack):
                duck.quack()
                
    
    def allOper(self):
        for duck in DuckManager.__duck_list:

            if isinstance(duck, Quack):
                duck.quack()
            if isinstance(duck, Fly):
                duck.fly()                
            if duck != None:
                duck.display()
                duck.swim()
            
            
```

```python
class MallardDuck(Duck, Quack, Fly):
    
    def __init__(self):
        super(MallardDuck, self).__init__()
        self.__myShape = "MallardDuck"
        self.__sound = "꽥꽥"
        
    def display(self):
        self._turtle = turtle
        self._turtle.color("blue")
        self._turtle.begin_fill()
        self._turtle.circle(Duck.DUCK_SIZE)
        self._turtle.end_fill()
        self._turtle.penup()
        self._turtle.setpos(self._x, self._y)
        self._turtle.pendown()
        print(self._x, ",", self._y, ":", self.__myShape)

    def quack(self):
        self._turtle = turtle
        self._turtle.color("black")
        self._turtle.penup()
        self._turtle.setpos(self._x + 40, self._y + 40)
        self._turtle.write(self.__sound)  #print text
        self._turtle.setpos(self._x, self._y)
        self._turtle.pendown()
        

       
        
class RedDuck(Duck, Quack, Fly):
    
    
    def __init__(self):
        super(RedDuck, self).__init__()
        self.__myShape = "RedDuck"
        self.__sound = "꽥꽥"
        
    def display(self):
        self._turtle = turtle
        self._turtle.color("red")
        self._turtle.begin_fill()
        self._turtle.circle(Duck.DUCK_SIZE)
        self._turtle.end_fill()
        self._turtle.penup()
        self._turtle.setpos(self._x, self._y)
        self._turtle.pendown()
        print(self._x, ",", self._y, ":", self.__myShape)
        
    def quack(self):
        self._turtle = turtle
        self._turtle.color("black")
        self._turtle.penup()
        self._turtle.setpos(self._x + 40, self._y+40)
        self._turtle.write(self.__sound)  #print text
        self._turtle.setpos(self._x, self._y)
        self._turtle.pendown()
   
        

class RubberDuck(Duck, Quack):
    
    def __init__(self):
        super(RubberDuck, self).__init__()
        self.__myShape = "RubberDuck"
        self.__sound = "삑삑"
        
    def display(self):
        self._turtle = turtle
        self._turtle.color("yellow")
        self._turtle.begin_fill()
        self._turtle.circle(Duck.DUCK_SIZE)
        self._turtle.end_fill()
        self._turtle.penup()
        self._turtle.goto(self._x, self._y)
        self._turtle.pendown()
        print(self._x, ",", self._y, ":", self.__myShape)
        
    def quack(self):
        self._turtle = turtle
        self._turtle.color("red")
        self._turtle.penup()
        self._turtle.setpos(self._x + 40, self._y+40)
        self._turtle.write(self.__sound)  #print text
        self._turtle.setpos(self._x, self._y)
        self._turtle.pendown()
        
        
class DecoyDuck(Duck):
    
    def __init__(self):
        super(DecoyDuck, self).__init__()
        self.__myShape = "DecoyDuck"
        
    def display(self):
        self._turtle = turtle
        self._turtle.color("green")
        self._turtle.begin_fill()
        self._turtle.circle(Duck.DUCK_SIZE)
        self._turtle.end_fill()
        self._turtle.penup()
        self._turtle.setpos(self._x, self._y)
        self._turtle.pendown()
        print(self._x, ",", self._y, ":", self.__myShape)
        
```
