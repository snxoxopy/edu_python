```python
import requests                  #   HTML  , response
from bs4 import BeautifulSoup    # HTML  
 
#     BeautifulSoup  
response = requests.get('https://pythondojang.bitbucket.io/weather/observation/currentweather.html')
#response.headers
response.content # body
print(response.content)
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
soup
```

```python
table = soup.find('table',{'class':'table_develop3'})
```

```python
table
```

```python
data = []
for tr in table.find_all('tr'):
    tds = list(tr.find_all('td'))
    print(tds)
    
    for td in tds:
        if td.find('a'):
            point = td.find('a').text
            temperature = tds[5].text
            humidity = tds[9].text
            data.append([point, temperature, humidity])
```

```python
soup = BeautifulSoup(response.content, 'html.parser')

##content_weather > table > tbody > tr:nth-child(1) > td:nth-child(6)
table = soup.find('table', { 'class': 'table_develop3' })    # <table class="table_develop3"> 
data = []                            #    
for tr in table.find_all('tr'):      #  <tr>   (   )
    tds = list(tr.find_all('td'))    #  <td>    
                                     # (    )
    for td in tds:                   # <td>   (   )
        if td.find('a'):             # <td>  <a>  ( )
            point = td.find('a').text    # <a>    
            temperature = tds[5].text    # <td>    ( 5)  
            humidity = tds[9].text       # <td>    ( 9)  
            data.append([point, temperature, humidity])    # data  , ,  

data    # data .   print     
```

```python
with open('weather.csv', 'w') as file:    # weather.csv    
    file.write('point,temperature,humidity\n')                  #   
    for i in data:                                              # data 
        file.write('{0},{1},{2}\n'.format(i[0], i[1], i[2]))    # ,,   
```
