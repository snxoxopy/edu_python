```python
import requests                  # 웹 페이지의 HTML을 가져오는 모듈, response까지
from bs4 import BeautifulSoup    # HTML을 파싱하는 모듈
 
# 웹 페이지를 가져온 뒤 BeautifulSoup 객체로 만듦
response = requests.get('https://movie.daum.net/ranking/reservation')
#response.headers
response.content # body
soup = BeautifulSoup(response.content, 'html.parser')

#link_txt = soup.find('a', { 'class': 'link_txt' })    # <a class="link_txt">을 찾음
data = []                            # 데이터를 저장할 리스트 생성
for txt_movies in soup.find_all('a', { 'class': 'link_txt' }):      # 모든 <a> 태그를 찾아서 반복(각 지점의 데이터를 가져옴)
    title = txt_movies.text
    data.append([title])

for i in range(0,10):
    print(data[i])
```
