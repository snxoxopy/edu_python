```python
import random
import turtle
from abc import *



class Duck(metaclass=ABCMeta): 

    DUCK_SIZE = 30 # class
    cnt = 0
    

    def  __init__(self, turtle):
        self._x = random.randint(-300,300) #instance
        self._y = random.randint(-300,300)
        self._turtle = turtle
        
        
        
    @abstractmethod #abstract method           
    def display(self):
        Duck.cnt += 1
        pass

        
    def screen_reset(self):
        self._turtle.reset()


    def quack(self):
        #self._turtle.color('black')
        self._turtle.penup()
        self._turtle.setpos(self._x + 40, self._y+40)
       
        self._turtle.write("quack")  #print text
        self._turtle.setpos(self._x, self._y)
        self._turtle.pendown()
        
    
    def move(self):
        #self._turtle.color('black')
        print(self._x, ',', self._y, ':', Duck.DUCK_SIZE)  
        self._turtle.penup()
        self._turtle.setpos(self._x - 50, self._y+40)
        
        self._turtle.write("move")  #print text
        self._turtle.setpos(self._x, self._y)
        self._turtle.pendown()

        
class MallardDuck(Duck): #ingeritance
    
    
    
    def  __init__(self, turtle):
        super().__init__(turtle)
        #self.screen_reset()
        if Duck.cnt == 0: self._turtle.reset()
        self._turtle.color('blue')
        
        
        
    # Method's Override
    def display(self):
        self._turtle.color('blue')
        self._turtle.begin_fill()
        self._turtle.circle(Duck.DUCK_SIZE)
        self._turtle.end_fill()
        self._turtle.penup()
        self._turtle.goto(self._x, self._y)
        self._turtle.pendown()
        self._turtle.penup()
        self._turtle.goto(self._x+39, self._y)
        Duck.cnt += 1
        self._turtle.write(Duck.cnt) 
        self._turtle.pendown()
        print(self._x, ',', self._y, ':', Duck.DUCK_SIZE, Duck.cnt)
    
    

class RedDuck(Duck): #ingeritance
    def __init__(self, turtle):
        super().__init__(turtle)
        if Duck.cnt == 0: self._turtle.reset()
        self._turtle.color('red')
        
        
    # Method's Override
    def display(self):
        self._turtle.color('red')
        self._turtle.begin_fill()
        self._turtle.circle(Duck.DUCK_SIZE)
        self._turtle.end_fill()
        self._turtle.penup()
        self._turtle.goto(self._x, self._y)
        self._turtle.pendown()
        print(self._x, ',', self._y, ':', Duck.DUCK_SIZE)
        Duck.cnt += 1
        self._turtle.penup()
        self._turtle.goto(self._x+30, self._y)
        self._turtle.write(Duck.cnt) 
        self._turtle.pendown()
        
   
```
