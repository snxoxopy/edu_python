```python
import random
from abc import abstractmethod
from abc import ABC
from abc import ABCMeta
import turtle


class Duck(metaclass=ABCMeta):
    #cnt = 0
    
    __DUCK_SIZE = 30
    #__DUCK_CNT = 0
    
    def __init__(self, turtle):
        #protected
        self._x = random.randint(-300,300)
        self._y = random.randint(-300,300)
        self._turtle = turtle
        self.screen_reset()
        
    def swim(self):
        self._turtle.color('black')
        self._turtle.penup()
        self._turtle.setpos(self._x - 50, self._y+40)  
        self._turtle.write("수영")  #print text
        self._turtle.setpos(self._x, self._y)
        self._turtle.pendown()
        print("수영")
            
    
    @abstractmethod    
    def display(self):
        #Duck.__DUCK_CNT += 1
        pass
    
    def screen_reset(self):
        self._turtle.reset()
   
        
```

```python
class Quack:
    
    def __init__(self):
        self._sound = ""
    

    @abstractmethod   
    def quack(self):
        pass
    
    
class Fly:
    
    def __init__(self):
        pass
    

    @abstractmethod   
    def fly(self):
        pass
```

```python
class DuckManager:

    __duck_list = []
    __instance = None
    
    def __init__(self):
        #self.makeDucks()
        if not DuckManager.__instance:
            print('creation')
        else:
            print('already', self.getInstance())
            
    @classmethod
    def getInstance(cls):
        if not cls.__instance:
            cls.__instance = DuckManager()
        return cls.__instance

    def makeDucks(self, turtle):
        for i in range(0,10):
            type = random.randint(0,3)
            if type == 0:
                DuckManager.__duck_list.append(MallardDuck(turtle))
            elif type == 1:
                DuckManager.__duck_list.append(RedDuck(turtle))
            elif type == 2:
                DuckManager.__duck_list.append(RubberDuck(turtle))
            else:
                DuckManager.__duck_list.append(DecoyDuck(turtle))
        
    def displayAllDucks(self):
        for duck in DuckManager.__duck_list:
            duck.output()
            
            
```

```python
class MallardDuck(Duck, Quack, Fly):
    
    __DUCK_SIZE = 30 # class
    
    def __init__(self, turtle):
        super(MallardDuck, self).__init__(turtle)
        self.__myShape = "MallardDuck"
        
    def display(self):
        self._turtle.color('blue')
        self._turtle.begin_fill()
        self._turtle.circle(self.__DUCK_SIZE)
        self._turtle.end_fill()
        self._turtle.penup()
        self._turtle.setpos(self._x, self._y)
        self._turtle.pendown()
        print(self._x, ",", self._y, ":", self.__myShape)

    def quack(self):
        self._turtle.color('black')
        self._turtle.penup()
        self._turtle.setpos(self._x + 40, self._y + 40)
        self._turtle.write("꽥꽥")  #print text
        self._turtle.setpos(self._x, self._y)
        self._turtle.pendown()
        
    def fly(self):
        self._turtle.color('black')
        self._turtle.penup()
        self._turtle.setpos(self._x - 50, self._y-20)
        self._turtle.write("날다")  #print text
        self._turtle.setpos(self._x, self._y)
        self._turtle.pendown()
        print("날다")
        
    def output(self):
        self.swim()
        self.quack()
        self.fly()
        self.display()
        
        
class RedDuck(Duck, Quack, Fly):
    
    __DUCK_SIZE = 30 # class
    
    def __init__(self, turtle):
        super(RedDuck, self).__init__(turtle)
        self.__myShape = "RedDuck"
        
    def display(self):
        self._turtle.color('red')
        self._turtle.begin_fill()
        self._turtle.circle(self.__DUCK_SIZE)
        self._turtle.end_fill()
        self._turtle.penup()
        self._turtle.setpos(self._x, self._y)
        self._turtle.pendown()
        print(self._x, ",", self._y, ":", self.__myShape)
        
    def quack(self):
        self._turtle.color('black')
        self._turtle.penup()
        self._turtle.setpos(self._x + 40, self._y+40)
        self._turtle.write("꽥꽥")  #print text
        self._turtle.setpos(self._x, self._y)
        self._turtle.pendown()
   
    def fly(self):
        self._turtle.color('black')
        self._turtle.penup()
        self._turtle.setpos(self._x - 50, self._y-20)
        self._turtle.write("날다")  #print text
        self._turtle.setpos(self._x, self._y)
        self._turtle.pendown()
        print("날다")
        
    def output(self):
        self.swim()
        self.quack()
        self.fly()     
        self.display()

class RubberDuck(Duck, Quack):
    
    __DUCK_SIZE = 30 # class
    
    def __init__(self, turtle):
        super(RubberDuck, self).__init__(turtle)
        self.__myShape = "RubberDuck"
        
    def display(self):
        self._turtle.color('yellow')
        self._turtle.begin_fill()
        self._turtle.circle(self.__DUCK_SIZE)
        self._turtle.end_fill()
        self._turtle.penup()
        self._turtle.goto(self._x, self._y)
        self._turtle.pendown()
        print(self._x, ",", self._y, ":", self.__myShape)
        
    def quack(self):
        self._turtle.color('red')
        self._turtle.penup()
        self._turtle.setpos(self._x + 40, self._y+40)
        self._turtle.write("삑삑")  #print text
        self._turtle.setpos(self._x, self._y)
        self._turtle.pendown()
        
    def output(self):
        self.swim()
        self.quack()
        self.display()
        #self.fly()    

        
class DecoyDuck(Duck):
    
    __DUCK_SIZE = 30 # class
    
    def __init__(self, turtle):
        super(DecoyDuck, self).__init__(turtle)
        self.__myShape = "DecoyDuck"
        
    def display(self):
        self._turtle.color('green')
        self._turtle.begin_fill()
        self._turtle.circle(self.__DUCK_SIZE)
        self._turtle.end_fill()
        self._turtle.penup()
        self._turtle.setpos(self._x, self._y)
        self._turtle.pendown()
        print(self._x, ",", self._y, ":", self.__myShape)
        
    def output(self):
        self.swim()
        self.display()
```
